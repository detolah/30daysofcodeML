```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer

breast_cancer = load_breast_cancer()
```

```python
print(breast_cancer.keys())
```

```python
print(breast_cancer.DESCR)
```

```python
cancer_data = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
cancer_data.head(10)
```

```python
cancer_data['class'] = breast_cancer.target
```

```python
cancer_data.tail(10)
```

```python
corr = cancer_data.corr(method = 'spearman')
corr
```

```python
#Extract 10 features with the best correlation with the target alongside the target into a new dataframe.

columns = np.full((corr.shape[0],), True, dtype=bool)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.7:
            if columns[j]:
                columns[j] = False
selected_columns = cancer_data.columns[columns]
corre_cancer_data = cancer_data[selected_columns]
```

```python
corre_cancer_data.head()
```

```python
#Top 5 features

from sklearn.feature_selection import SelectKBest, chi2
y = corre_cancer_data['class']
Newcorre_cancer_data = SelectKBest(chi2, k=6).fit_transform(corre_cancer_data,y)
```

```python
Newcorre_cancer_data
```

```python
features = ['mean radius','mean texture','mean symmetry','mean compactness','radius error','class']
```

```python
cancer_topfeatures = pd.DataFrame(Newcorre_cancer_data, columns = features)
```

```python
cancer_topfeatures.head()
```

```python
#make a pair plot
sns.pairplot(cancer_topfeatures)
```

```python
cancer_topfeatures.mean()
```

```python
#Make a copy of this version of the dataframe and for each feature replace all values less than their respective mean with NaN.
cancer_topfeatures['mean radius'].loc[(cancer_topfeatures['mean radius'] < cancer_topfeatures['mean radius'].mean() )] = np.NaN
```

```python
cancer_topfeatures['mean radius'].loc[(cancer_topfeatures['mean texture'] < cancer_topfeatures['mean texture'].mean() )] =np.NaN
cancer_topfeatures['mean symmetry'].loc[(cancer_topfeatures['mean symmetry'] < cancer_topfeatures['mean symmetry'].mean() )] = np.NaN
cancer_topfeatures['mean compactness'].loc[(cancer_topfeatures['mean compactness'] < cancer_topfeatures['mean compactness'].mean() )] = np.NaN
cancer_topfeatures['radius error'].loc[(cancer_topfeatures['radius error'] < cancer_topfeatures['radius error'].mean() )] = np.NaN
```
