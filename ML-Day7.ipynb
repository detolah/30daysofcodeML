```python
#drop NAN datas
data = data.dropna()
```

```python
col1 = data['CO(GT)']
col2= data['C6H6(GT)']
col3 = data['T']
col4 = data['RH']
col5 = data['AH']
```

```python
data['CO(GT)'] = col1.str.replace(',', '.')
data['C6H6(GT)'] = col2.str.replace(',', '.')
data['RH'] = col4.str.replace(',', '.')
data['T'] = col3.str.replace(',', '.')
data['AH'] = col5.str.replace(',', '.')
```

```python
data['CO(GT)'] = data['CO(GT)'].astype(float)
data['RH'] = data['RH'].astype(float)
data['T'] = data['T'].astype(float)
data['C6H6(GT)'] = data['C6H6(GT)'].astype(float)
```

```python
data['Date'] = pd.to_datetime(data['Date'])

data['Days'] = data['Date'].dt.weekday_name
data['Months'] = data['Date'].dt.month
```

```python
data.head()
```

```python
def timeOfDay(x):
    if (x < 8):
        return 'Morning'
    elif (x >= 8) and (x < 16 ):
        return 'Afternoon'
    elif (x >= 16):
        return'Evening'
```

```python
extr = data['Time'].str.extract(r'^(\d{2})', expand=False)
timenum = pd.to_numeric(extr)
data['Hour of Day'] = timenum.apply(timeOfDay)
```
